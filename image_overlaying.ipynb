```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
zeb = cv2.imread("zebra.jpg")
disk = cv2.imread("recording.jpg")
zeb = cv2.cvtColor(zeb,cv2.COLOR_BGR2GRAY)
disk = cv2.cvtColor(disk,cv2.COLOR_BGR2GRAY)
```

```python
thresh = 150
max = 255
th, dst = cv2.threshold(disk,thresh,max,cv2.THRESH_BINARY)
```

```python
plt.imshow(dst[:-1],cmap='gray')
```

```python
plt.imshow(zeb[:-1],cmap='gray')
```
